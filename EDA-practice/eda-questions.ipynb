## 14) What is the total coffee production for each year?

```python
df.groupby('Year')['Total Production'].sum()
```

```python
years = df['Year'].unique()
prod_per_year = {}

for year in years:
    prod_per_year[year] = np.sum(df[df['Year'] == year]['Total Production'])

pd.DataFrame({
    'Year': prod_per_year.keys(),
    'Total Production': prod_per_year.values()
})
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
plt.bar(years, prod_per_year.values())
plt.xlabel('Year')
plt.ylabel('Total Production')
plt.xticks(years)
plt.title('Total Coffee Production Per Year')
plt.show()
```

## 18) Which country shows the highest growth in total production over the years?

```python
countries = df['Country'].unique()
years = df['Year'].unique()

for country in countries:
    country_df = df[df['Country'] == country]

    prod_per_yr = {}
    for year in years:
        prod_per_yr[year] = country_df[country_df['Year'] == year]['Total Production'].sum()

    plt.figure(figsize=(2,2))
    plt.barh(years, prod_per_yr.values())
    plt.ylabel('Year')
    plt.xlabel('Total Production')
    plt.yticks(years)
    plt.title(f'Country: {country}')
    plt.show()
```

```python
countries = df['Country'].unique()
years = df['Year'].unique()

growth_per_country = {}

for country in countries:
    country_df = df[df['Country'] == country]

    prod_per_yr = {}
    for year in years:
        prod_per_yr[year] = country_df[country_df['Year'] == year]['Total Production'].sum()
    
    final_prod = prod_per_yr[years[-1]]
    start_prod = prod_per_yr[years[0]]

    growth_per_country[country] = (final_prod - start_prod)


plt.figure(figsize=(8,6))
plt.barh(growth_per_country.keys(), growth_per_country.values())
plt.ylabel('Country')
plt.xlabel('Total Growth in Production')
plt.title(f'Total growth in total production over the years')
plt.show()
```

```python
max_growth = max(growth_per_country.values())
for country in countries:
    if growth_per_country[country] == max_growth:
        print(country)
```

```python
# China shows the highest growth in total production over the years
```
